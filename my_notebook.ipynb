```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', as_frame=False)
X, y = mnist.data, mnist.target
```

```python
import matplotlib.pyplot as plt
```

```python
arr = X[1].reshape(28,28)
```

```python
arr = X[24].reshape(28,28)

plt.imshow(arr, cmap='gray')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.rc('font', size=9)  
plt.imshow(arr)

for i in range(0, arr.shape[0]):
    for j in range(0, arr.shape[1]):
        plt.text(j, i, int(arr[i,j]), horizontalalignment="center", verticalalignment="center")
        
plt.show()
```

# Normalization

```python
X = arr.copy()
X_normalized = (X - X.min()) / (X.max() - X.min()) * 255

plt.figure(figsize=(10,10))
plt.rc('font', size=6)

plt.imshow(X_normalized)

for i in range(0, X_normalized.shape[0]):
    for j in range(0, X_normalized.shape[1]):
        plt.text(j, i, float(X_normalized[i,j].round(1)), \
                 horizontalalignment="center", verticalalignment="center")

plt.show()
```
